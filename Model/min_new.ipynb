```python
import random
import numpy as np
import pandas as pd
from itertools import combinations
def rnd_pwl(xmin, xmax, g, size=1):
    r = np.random.random(size=size)
    return (r * (xmax ** (1. - g) - xmin ** (1. - g)) + xmin ** (1. - g)) ** (1. / (1. - g))

def HOADmodel(N: int, activities_per_order: dict, time = 100):
    hyperlinks = []
    time_a = []
    prob_matrix = np.ones(N) - np.identity(N)
    for t in range(time):
        for order in activities_per_order.keys():
            act_vect = activities_per_order[order]
            for node_i in range(N):
                if act_vect[node_i] > random.random():
                    neigh_list = list(np.random.choice(range(N), order, p = prob_matrix[node_i]/sum(prob_matrix[node_i])))
                    neigh_list.append(node_i)
                    if len(neigh_list) == len(set(neigh_list)):
                        for pair in list(combinations(neigh_list,2)):
                            prob_matrix[pair] += 1
                        hyperlinks.append(neigh_list)
                        time_a.append(t)
    return [time_a, hyperlinks]
```

```python
orders = [2,3,4,5,6,7,8,9,10]
N = 1000

activities_per_order ={}
activity_list = sorted([rnd_pwl(.001, 1, 2.25)[0] for i in range(N)])
for order in orders:
    activities_per_order[order]= activity_list

t, simplices = HOADmodel(N,activities_per_order)
```
